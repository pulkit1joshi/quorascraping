```python
k = 0
while k < 20:
        from selenium import webdriver
        from selenium.webdriver.common.by import By
        from bs4 import BeautifulSoup
        from unidecode import unidecode 
        import time
        import pandas as pd
        df = pd.read_csv("data.csv")
        df.drop_duplicates(subset=None, keep="first", inplace=True)

        sleep = 10
        USER_NAME = '' #Insert Account name here
        PASS_WORD = '' #Insert Account Password here
        url = 'https://www.quora.com'
        url2 = ['https://www.quora.com']
        #Logging in to your account
        driver = webdriver.Firefox(executable_path=r'C:\Users\Pulkit Joshi\Downloads\Compressed\geckodriver-v0.24.0-win64\geckodriver.exe')
        driver.get(url)
        page_source=driver.page_source
        username = driver.find_element(By.XPATH, '//input[@placeholder="Email"]')
        password = driver.find_element(By.XPATH, '//input[@placeholder="Password"]')
        login= driver.find_element(By.XPATH, '//input[@value="Login"]')
        username.send_keys(USER_NAME)
        password.send_keys(PASS_WORD)
        time.sleep(sleep)
        login.click()
        time.sleep(sleep)
        driver.get("https://www.quora.com/search?q=interview")
        source = driver.page_source
        soup = BeautifulSoup(source,"html.parser")
        urls = [x.get('href') for x in soup.findAll("a", class_ = "question_link")]
        flag = 0;
        while len(urls) < 2000 :
            lasturl = len(urls)
            source = driver.page_source
            soup = BeautifulSoup(source,"html.parser")
            print(len(urls))
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            urls = [x.get('href') for x in soup.findAll("a", class_ = "question_link")]
            time.sleep(1)
            if lasturl == len(urls):
                flag = flag+1;
                print(flag)
            if flag == 10 :
                break
        urls
        def turntourl(url):
            # remove the index.html part of the base url before returning the results
            y = "https://www.quora.com" + url;
            return(y)
        prop_urls = [];
        for x in urls:
            prop_urls.append(turntourl(x))
        def getfollow(soup):
            # remove the index.html part of the base url before returning the results
                x = 0;
                i=0;
                if(soup.find("span" , class_ = "ui_button_count_inner")):
                    y = soup.find("span" , class_ = "ui_button_count_inner").text
                    while i <len(y):
                        if y[i]=='k': 
                            x=x*100;
                        elif y[i]=='.':
                            i = i
                        else:
                            x =x*10+ int(y[i]) 
                        i = i + 1
                    return(x)
                else:
                    return(0)

        def getans(soup):
                x = 0
                if(soup.find("div" , class_ = "answer_count")):
                    y = soup.find("div" , class_ = "answer_count").text
                    i = 0
                    while i<len(y):
                        if y[i] >= '0' and y[i] <= '9':
                            x = x*10 + int(y[i])
                        i=i+1
                    return x
                else:
                    return(0);
        def getid(soup):
            
            i=0
            x=0
            if(soup.find("a" , class_ = "ui_button u-nowrap ui_button--styled ui_button--FlatStyle ui_button--FlatStyle--gray ui_button--size_regular u-inline-block ui_button--non_link ui_button--supports_icon ui_button--has_icon")):
                if(soup.find("a" , class_ = "ui_button u-nowrap ui_button--styled ui_button--FlatStyle ui_button--FlatStyle--gray ui_button--size_regular u-inline-block ui_button--non_link ui_button--supports_icon ui_button--has_icon").has_attr('action_target')):
                    y = soup.find("a" , class_ = "ui_button u-nowrap ui_button--styled ui_button--FlatStyle ui_button--FlatStyle--gray ui_button--size_regular u-inline-block ui_button--non_link ui_button--supports_icon ui_button--has_icon").get("action_target")
                    while i<len(y):
                        if y[i] >= '0' and y[i] <= '9':
                            x = x*10 + int(y[i])
                        i=i+1
                    return x
                else:
                    return(0)
            else:
                return(0)

        def getques(soup):
            y = soup.find("span" , class_ = "rendered_qtext").text
            return y
        
        def extract(y):
            i=0
            x=''
            while i<len(y):
                if(y[i] == 'c' and y[i+1] =='o' and y[i+2] == 'm' and y[i+3] == '/') :
                    i=i+4
                    break
                i=i+1
                    
            while i<len(y):
                x = x + y[i]
                i=i+1
            return x
        
        i = 0
        for x in prop_urls:

            print(i)
            driver.get(x)
            source = driver.page_source
            soup = BeautifulSoup(source,"html.parser")
            follow = getfollow(soup)
            follow = int(follow);
            ans = getans(soup);
            ans = int(ans)
            qid = getid(soup)
            print(follow)
            que = extract(x)
                
            print(extract(x))
            if(follow > 39 and ans >9):
                df = df.append({'QID': qid, 'URL': x, 'Question': que , 'Follows': follow, 'Answers': ans}, ignore_index=True)
            i=i+1;
        df.head()
        df
        df.drop_duplicates(subset=None, keep="first", inplace=True)
        df.to_csv('data.csv' , index = False)
        df
        k=k+1
```
